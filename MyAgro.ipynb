Before I solve this question let us make a clear remark and assumption because the Mentor and mentee is not clearly defined.

*   **Mentors:** Every female who is at least 40yo (>=40) and who joined the community before or on the 1st January 2004 (<= 2004-01-01).

*   **Mentors:** Every male who is at least 40yo (>=40) and who joined the community before or on the 1st January 2004 (<= 2004-01-01).

*   Else you are considered a **Mentee**


```python
mentoring = []
date_regis = []
FullName = []
Age = []
Mentorship = pd.DataFrame()
for i in range(0, 500):
  constant_date = pd.to_datetime('2004-01-01').date()
  if CommunityMember.loc[i, 'Gender'] == "female" and CommunityMember.loc[i, 'Age'] >= 40 and  CommunityMember.loc[i, 'Registered']['date'] == constant_date:
    mentoring.append("Mentor")
    date = pd.to_datetime(CommunityMember.loc[i, 'Registered']['date']).date()
    date_regis.append(date)
    FullName.append(CommunityMember.loc[i, 'FullName'])
    Age.append(CommunityMember.loc[i, 'Age'])
  elif CommunityMember.loc[i, 'Gender'] == "male" and CommunityMember.loc[i, 'Age'] >= 40 and  CommunityMember.loc[i, 'Registered']['date'] == constant_date:
    mentoring.append("Mentor")
    date = pd.to_datetime(CommunityMember.loc[i, 'Registered']['date']).date()
    date_regis.append(date)
    FullName.append(CommunityMember.loc[i, 'FullName'])
    Age.append(CommunityMember.loc[i, 'Age'])
  elif CommunityMember.loc[i, 'Age'] <= 30:
    mentoring.append("Mentee")
    date = pd.to_datetime(CommunityMember.loc[i, 'Registered']['date']).date()
    date_regis.append(date)
    FullName.append(CommunityMember.loc[i, 'FullName'])
    Age.append(CommunityMember.loc[i, 'Age'])
  else:
    mentoring.append("Mentee")
    date = pd.to_datetime(CommunityMember.loc[i, 'Registered']['date']).date()
    date_regis.append(date)
    FullName.append(CommunityMember.loc[i, 'FullName'])
    Age.append(CommunityMember.loc[i, 'Age'])
Mentorship["FullName"] = FullName
Mentorship["Mentoring"] = mentoring
Mentorship["Registration Date"] = date_regis
Mentorship["Age"] = Age
Mentorship.head()
```

**Step 8:** Create a function that will take as parameters the 2 dataframes created in Step 6&7 and return the **median age of male mentorees that registered after 2010 (>=2010-01-01).**


```python
def medianAgeMaleMentee(CommunityMember, Mentorship):
  constant_date = pd.to_datetime('2010-01-01').date()
  Mentorship['Gender'] = CommunityMember['Gender']
  Mentorship_2010 = Mentorship[Mentorship['Registration Date'] >= constant_date]
  Mentorship_2010_Male = Mentorship_2010[Mentorship_2010['Gender'] == 'male']
  return Mentorship_2010_Male.median()
```
